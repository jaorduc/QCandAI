<div id="header"><p style="color:#2ea44f; text-align:center; font-weight:bold; font-family:verdana; font-size:18px;">Part I: Puting a number in a Quantum Circuit and implementing a quantum simulator</p></div>

````python
from qiskit import QuantumRegister, QuantumCircuit, ClassicalRegister
import qiskit
import matplotlib.pyplot as plt
%matplotlib inline
````

<div id="header"><p style="color:#2ea44f; text-align:left; font-weight:bold; font-family:verdana; font-size:14px;">Create a qubit register</p></div>

````python
qa = QuantumRegister(3, name="qa")
qb = QuantumRegister(3,name="qb")
cc = ClassicalRegister(3,name="cc")
qc = QuantumCircuit(qa, qb, cc)
````

````python
qc.draw(output="mpl")
````

## 


````python
qc.x(qa[0])
qc.x(qa[2])
qc.draw(output="mpl")
````

````python
qc.measure(qa, cc)
qc.draw(output="mpl")
````

````python
from qiskit_aer import Aer, AerSimulator
````

````python
bcknd = Aer.get_backend('qasm_simulator')
````

````python
from qiskit import transpile
````

Instead using 
```python 
job = execute(circuit, backend)
``` 
we should use 
```python 
qctranspile = transpile(qc, bcknd)
job = bcknd.run(qctranspile)
```
as follows

````python
qctranspile = transpile(qc, bcknd)
job = bcknd.run(qctranspile)
````

````python
result = job.result()
counts = result.get_counts(qc)
print(counts)
````
